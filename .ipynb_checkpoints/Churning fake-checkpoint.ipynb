```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Initialize Faker for realistic fake data
fake = Faker()

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)
Faker.seed(42)

# Define parameters
num_customers = 10000  # Number of records to generate

## Generate base customer data
data = {
    'CLIENTNUM': [f"CID-{str(i).zfill(5)}" for i in range(1, num_customers+1)],
    'Attrition_Flag': np.random.choice(['Existing', 'Attrited'], size=num_customers, p=[0.85, 0.15]),
    'Customer_Age': np.random.randint(18, 70, size=num_customers),
    'Gender': np.random.choice(['M', 'F'], size=num_customers, p=[0.6, 0.4]),
    'Dependent_count': np.random.randint(0, 6, size=num_customers),
    'Education_Level': np.random.choice(
        ['High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate', 'Uneducated', 'Unknown'],
        size=num_customers,
        p=[0.2, 0.15, 0.3, 0.1, 0.05, 0.1, 0.1]
    ),
    'Marital_Status': np.random.choice(
        ['Single', 'Married', 'Divorced', 'Unknown'],
        size=num_customers,
        p=[0.3, 0.5, 0.15, 0.05]
    ),
    'Income_Category': np.random.choice(
        ['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Unknown'],
        size=num_customers,
        p=[0.15, 0.25, 0.2, 0.15, 0.15, 0.1]
    ),
    'Card_Category': np.random.choice(
        ['Blue', 'Silver', 'Gold', 'Platinum'],
        size=num_customers,
        p=[0.7, 0.2, 0.08, 0.02]
    )
}

## Generate numerical financial metrics with correlations
# Base values that will correlate with other features
months_on_book = np.random.randint(12, 60, size=num_customers)
credit_limit = np.random.normal(5000, 2000, size=num_customers).astype(int)

# Generate correlated features
data.update({
    'Months_on_Book': months_on_book,
    'Total_Relationship_Count': np.random.randint(1, 7, size=num_customers),
    'Months_Inactive_12_mon': np.random.poisson(2, size=num_customers),
    'Contacts_Count_12_mon': np.random.poisson(3, size=num_customers),
    'Credit_Limit': np.clip(credit_limit, 1000, 30000),
    'Total_Revolving_Bal': np.random.normal(1500, 800, size=num_customers).astype(int),
    'Avg_Open_To_Buy': np.clip(credit_limit - np.abs(np.random.normal(1000, 500, size=num_customers)), 0, None),
    'Total_Amt_Chng_Q4_Q1': np.random.uniform(0.5, 3.5, size=num_customers),
    'Total_Trans_Amt': np.random.normal(4000, 2000, size=num_customers).astype(int),
    'Total_Trans_Ct': np.random.randint(10, 100, size=num_customers),
    'Total_Ct_Chng_Q4_Q1': np.random.uniform(0.5, 3.5, size=num_customers),
    'Avg_Utilization_Ratio': np.random.uniform(0, 0.85, size=num_customers)
})

# Create DataFrame
df = pd.DataFrame(data)

# Add some realistic relationships between variables
# Higher income customers tend to have higher credit limits
df.loc[df['Income_Category'].isin(['$80K - $120K', '$120K +']), 'Credit_Limit'] *= 1.5
df.loc[df['Income_Category'] == 'Less than $40K', 'Credit_Limit'] *= 0.7

# Platinum card holders have higher limits
df.loc[df['Card_Category'] == 'Platinum', 'Credit_Limit'] *= 2
df.loc[df['Card_Category'] == 'Gold', 'Credit_Limit'] *= 1.5

# Attrited customers tend to have higher inactivity
df.loc[df['Attrition_Flag'] == 'Attrited', 'Months_Inactive_12_mon'] += 1
df.loc[df['Attrition_Flag'] == 'Attrited', 'Contacts_Count_12_mon'] += 1

# Add some missing values realistically
for col in ['Education_Level', 'Marital_Status', 'Income_Category']:
    df.loc[df.sample(frac=0.05).index, col] = 'Unknown'

# Save to CSV - using a temporary directory that should have write permissions
try:
    output_path = os.path.join(os.getcwd(), 'synthetic_customer_attrition_data.csv')
    df.to_csv(output_path, index=False)
    print(f"File successfully saved to: {output_path}")
except PermissionError:
    # If still having permission issues, try the user's home directory
    home_dir = os.path.expanduser('~')
    output_path = os.path.join(home_dir, 'synthetic_customer_attrition_data.csv')
    df.to_csv(output_path, index=False)
    print(f"File saved to your home directory: {output_path}")

print("\nSynthetic dataset generated with shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nColumn names:")
print(df.columns.tolist())
```

```python
copy_df=df.copy()
copy_df.drop(columns=['Dependent_count'],inplace=True)
copy_df
```

```python


# Identify categorical columns
categorical_cols = copy_df.select_dtypes(include=['object', 'category']).columns

# Label encode categorical variables
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    copy_df[col] = le.fit_transform(copy_df[col].astype(str))
    label_encoders[col] = le

# Calculate correlation matrix
corr_matrix = copy_df.corr(method='pearson')  

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Generate a heatmap
sns.heatmap(corr_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            center=0,
            linewidths=0.5,
            cbar_kws={"shrink": 0.8})

# Adjust layout
plt.title('Correlation Matrix Heatmap', pad=20, fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```
