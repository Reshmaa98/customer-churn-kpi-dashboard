```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
import xgboost as xgb
import joblib
import streamlit as st
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

```

```python
#Loading the data and handling missing values
```

```python
churn_df=pd.read_csv("BankChurners.csv")
print("The number of rows and columns in the dataset is:",churn_df.shape)
print("The top five datapoints of the dataset is:")
print(churn_df.head(5))
print("The bottom five datapoints of the dataset is:")
print(churn_df.tail(5))
print(churn_df.dtypes)
churn_null=churn_df.isnull().sum()
print(churn_null)
for cols in churn_df.columns:
    if cols=='Customer_Age':
         churn_df[cols].fillna(churn_df[cols].mode()[0], inplace=True)
    elif churn_df[cols].dtype == "object":  
        churn_df[cols].fillna(churn_df[cols].mode()[0], inplace=True)  
    else:  
        churn_df[cols].fillna(churn_df[cols].mean(), inplace=True) 
    
```

```python
churn_copy_df=churn_df.copy()
churn_copy_df.drop(columns=['Dependent_count','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
                            'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'],inplace=True)
```

```python
print("The summary statistics of the dataset:")
print(churn_copy_df.describe().round(2))
```

```python
print(churn_copy_df.head(5))
print("The number of datapoints is:",churn_copy_df.shape)
```

```python
# Removing Outliers
```

```python
sns.boxplot(churn_copy_df['Total_Trans_Ct'])
plt.title("Boxplot of Total Transaction Count")
plt.show()
```

```python
Q1 = churn_copy_df["Total_Trans_Ct"].quantile(0.25)
Q3 = churn_copy_df["Total_Trans_Ct"].quantile(0.75)
IQR = Q3 - Q1

# Upper bound for outliers
upper_bound = Q3 + 1.5 * IQR

# Remove all upper outliers
churn_copy_df = churn_copy_df[churn_copy_df["Total_Trans_Ct"] <= upper_bound]

# Plot to verify
plt.figure(figsize=(6, 4))
sns.boxplot(y=churn_copy_df["Total_Trans_Ct"])
plt.show()
```

```python
sns.boxplot(churn_copy_df['Total_Trans_Amt'])
plt.title("Boxplot of Total Transaction Amount")
plt.show()
```

```python
threshold = churn_copy_df['Total_Trans_Amt'].quantile(0.98)
churn_copy_df = churn_copy_df[churn_copy_df['Total_Trans_Amt'] <= threshold]
sns.boxplot(churn_copy_df['Total_Trans_Amt'])
plt.title('Filtered (Top 2% Removed)')
plt.show()
```

```python
print("The number of datapoints after removing the outliers:",churn_copy_df.shape)
```

```python
# Building a model
```

```python
copy_df=churn_copy_df.copy()
copy_df.drop(columns="CLIENTNUM",inplace=True)
```

```python
# Building ML Model using XGBoost
```

```python
category_columns=['Attrition_Flag','Gender','Education_Level','Marital_Status','Income_Category','Card_Category']
label_encoders={}
for col in category_columns:
    le=LabelEncoder()
    copy_df[col]=le.fit_transform(copy_df[col])
    label_encoders[col] = le
X=copy_df.iloc[:,1:19]
Y=copy_df["Attrition_Flag"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=40)

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss")
xgb_model.fit(X_train, Y_train)





```

```python

# Generate predictions
Y_pred = xgb_model.predict(X_test)


# Calculate accuracy
accuracy = accuracy_score(Y_test, Y_pred)

# Print all metrics
print("=== Model Evaluation ===")
print(f"\nAccuracy: {accuracy:.2f}")  
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred, target_names=['Non-Churn', 'Churn'], digits=4))

print("\nConfusion Matrix:")
print(confusion_matrix(Y_test, Y_pred))
joblib.dump(xgb_model,"xgb_churn_model.pkl")


```
